```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
data = pd.read_csv('KM_plot_plain.csv')  # Replace with your actual file path

# Plot the Kaplan-Meier curves with 95% CI for the two cohorts
plt.figure(figsize=(10, 6))

# Plot for Cohort 1 with purple color and its 95% CI
plt.step(data['Time (Days)'], data['Cohort 1: Survival Probability'], where="post", label='Cohort 1: CRP > 2 mg/ml', color='purple')
plt.fill_between(data['Time (Days)'], data['Cohort 1: Survival Probability 95 % CI Lower'], data['Cohort 1: Survival Probability 95 % CI Upper'], color='purple', alpha=0.3)

# Plot for Cohort 2 with green color and its 95% CI
plt.step(data['Time (Days)'], data['Cohort 2: Survival Probability'], where="post", label='Cohort 2: CRP < 2 mg/ml', color='green')
plt.fill_between(data['Time (Days)'], data['Cohort 2: Survival Probability 95 % CI Lower'], data['Cohort 2: Survival Probability 95 % CI Upper'], color='green', alpha=0.3)

# Set labels and title
plt.xlabel('Time (Days)')
plt.ylabel('Survival Probability')
plt.title('Kaplan-Meier Survival Curves with 95% CI')
plt.ylim(0.96, 1)  # Restrict Y range between 0.96-1
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
